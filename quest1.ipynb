```python
word = str(input('입력값 :'))
reverse_name = ""
for c in word:
  reverse_name = c+ reverse_name
print('출력값:'+reverse_name)
print('뒤집힌 단어는:' +reverse_name)

if reverse_name == word:
  print('입력된 값은 회문입니다')
else:
  print('입력된 값은 회문이 아닙니다')

```
